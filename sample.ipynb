```python
def editDistance(word1, word2, i, j):
    if not word1 and not word2: 
        return 0
    if len(word1) == i:
        return len(word2) - j
    if len(word2) == j:
        return len(word1) - i
    if word1[i] == word2[j]:
        return editDistance(word1, word2, i+1, j+1)
    insert = editDistance(word1, word2, i, j+1)
    delete = editDistance(word1, word2, i+1, j)
    replace = editDistance(word1, word2, i+1, j+1)
    return min(insert, replace, delete) + 1
```

```python
def editDistanceTopDown(word1, word2, i, j, dp):
    tmp = 0;
    if not word1 and not word2: 
        return 0
    if len(word1) == i:
        return len(word2) - j
    if len(word2) == j:
        return len(word1) - i
    if dp[i][j] > 0:
        return dp[i][j]
    if word1[i] == word2[j]:
        tmp = editDistanceTopDown(word1, word2, i+1, j+1, dp)
    else:
        insert = editDistanceTopDown(word1, word2, i, j+1, dp)
        delete = editDistanceTopDown(word1, word2, i+1, j, dp)
        replace = editDistanceTopDown(word1, word2, i+1, j+1, dp)
        tmp = min(insert, replace, delete) + 1
    dp[i][j] = tmp
    return tmp
```

```python
def editDistanceBottomUp(word1, word2):

    m = len(word1)
    n = len(word2)
    dp = [[0 for x in range(n+1)] for y in range(m+1)] 
    for i in range(1, m+1):
        dp[i][0] = i;
    for j in range(1, n+1):
        dp[0][j] = j;
    for i in range(1, m+1):
        for j in range(1, n+1):
            if word1[i-1] == word2[j-1]:
                dp[i][j] = dp[i-1][j-1];
            else:
                dp[i][j] = min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1]) + 1
    return dp[m][n]
```

```python
def word2vec(word):
    from collections import Counter
    from math import sqrt

    # count the characters in word
    cw = Counter(word)
    # precomputes a set of the different characters
    sw = set(cw)
    # precomputes the "length" of the word vector
    lw = sqrt(sum(c*c for c in cw.values()))

    # return a tuple
    return cw, sw, lw

def cosdis(v1, v2):
    common = v1[1].intersection(v2[1])
    # by definition of cosine distance we have
    return sum(v1[0][ch]*v2[0][ch] for ch in common)/v1[2]/v2[2]
```

```python
testa = 'ASDASDFEWEGEGEGE'
testb = 'AASFDSFDSASHEHEEGE'
va = word2vec(testa)
vb = word2vec(testb)
print(cosdis(va,vb))
```
